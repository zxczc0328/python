```python
a=10
b=20
try:
    if a<b:
        print(n)
except:
    print("錯誤產生")
```

### 指定錯誤類別處理

```python
a=10
b=20
try:
    if a<b:
        print(n)
except NameError as e:
    print("錯誤產生, 錯誤原因：",e)
except ZeroDivisionError:
    print("除以0錯誤")
```

### else
##### 沒有錯誤產生會執行的部分

```python
a=10
b=20
try:
    if a>b: #改變比較條件，讓錯誤不發生
        print(n)
except:
    print("錯誤發生")
else:
    print("沒有錯誤發生時會執行的部分")
```

```python
a=10
b=20
try:
    if a>b: #改變比較條件，讓錯誤不發生
        print(n)
except:
    print("錯誤發生")
else:
    print("沒有錯誤發生時會執行的部分")
finally:
    print("不管有沒有錯誤一定會執行的部分")
```
